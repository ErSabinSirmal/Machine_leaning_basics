## Univeriate anylysis

```python
df['quality'].nunique() # what are the unique values of the given column?
```

```python
quality_counts = df['quality'].value_counts()
print(quality_counts)
quality_counts.index
```

#### Bar plot for evaluating the count of the wine with its quality rate

```python
plt.figure(figsize=(8,6))
plt.bar(quality_counts.index, quality_counts, color='skyblue')
plt.title('Count plot of quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

 ## Univariate Analysis with campus X

```python
df = pd.read_csv('Datasets/Titanic-Dataset.csv')
df.head()
```

## categorical data = ['Survived', 'Pclass', 'Sex', 'SibSp', 'Cabin', 'Embarked']

 ### a. count plot

```python
sns.countplot(x = df['Survived'], color = 'red')
```

```python
sns.countplot(x = df['Pclass'])
```

```python
df['Pclass'].value_counts()
```

 ### Pie chart

```python
plt.pie(df['Embarked'].value_counts(), 
        labels = df['Embarked'].value_counts().index,
        autopct = '%.2f%%')
plt.title("Pie chart with percentage")
plt.show()
df['Pclass'].value_counts().index
```

## Numerical Data
