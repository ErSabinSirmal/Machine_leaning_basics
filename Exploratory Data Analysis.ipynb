## Bivariate Analysis

#### Cross tabulation 

```python
titanic_df = pd.read_csv('Titanic-Dataset.csv')
titanic_df.head()
```

#### Identify categorical and target variables

```python
categorical_features = ['Pclass', 'Sex']
target_variable = 'Survived'
for col in categorical_features + [target_variable]:
    print(f"{col} unique values: ", titanic_df[col].unique())
```

#### Cross-Tabulation for Passenger Class vs Survival

```python
cross_tab_Pclass = pd.crosstab(
    titanic_df['Pclass'],
    titanic_df['Survived'],
    margins = True
)
cross_tab_Pclass
```

#### Cross-Tabulation for Gender vs Survival

```python
cross_tab_gender = pd.crosstab(
    titanic_df['Sex'],
    titanic_df['Survived'],
    margins = True
)
cross_tab_gender
```

#### Caculate survival percentages

```python
survival_rate_gender = pd.crosstab(
    titanic_df['Sex'],
    titanic_df['Survived'],
    normalize = True
) * 100

survival_rate_gender
```
