## Standardization

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Datasets/Social_Network_Ads.csv')
df.head()
```

```python
df1 = df.drop(columns = ['User ID', 'Gender']) #df = df.iloc[:,2:] same 
#df.columns
df1.head()
```

### Train Test Split

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
train_df = df1.drop('Purchased', axis=1) 
X_train, X_test, y_train, y_test = train_test_split(train_df,
                                                    df1['Purchased'],
                                                    test_size=0.3,
                                                    random_state=0)
X_train.shape, X_test.shape

```

### Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X_train) # Fit the scaler to the train set, it will learn the parameters

# Transnform train and test sets.
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
```

### Scaler takes Dataframe but return the numpy array so to view scaled data we can convert the ndarray into dataframe.

```python
X_train
```

```python
y_train
```

```python
X_train_scaled # which is numpy array.
```

### We can represent this numpy array in dataframe using this technique.

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_train_scaled
```

```python
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_train.columns)
X_test_scaled
```

```python
np.round(X_train.describe(), 1) # Before Scaling
```

```python
np.round(X_train_scaled.describe(), 1) # After scaling.
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12, 5))

ax1.scatter(x = X_train['Age'], y = X_train['EstimatedSalary'])
ax1.set_title('Before Scaling')
ax2.scatter(x = X_train_scaled['Age'], y = X_train_scaled['EstimatedSalary'], color = 'red')
ax2.set_title('After Scaling')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12,5))
# Before Scaling
ax1.set_title('Before scaling')
sns.kdeplot(x = X_train['Age'], ax = ax1, label = 'Age')
sns.kdeplot(x = X_train['EstimatedSalary'], ax = ax1, label = 'EstimatedSalary')
ax1.legend()
ax2.set_title('After scaling')
sns.kdeplot(x = X_train_scaled['Age'], ax = ax2, label = 'Age')
sns.kdeplot(x = X_train_scaled['EstimatedSalary'], ax = ax2, label = 'EstimatedSalary')
ax2.legend() # help to display the box represents the each label with different color.
plt.show()
```

### Comparision of Distribution

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12,5))

ax1.set_title('Age distribution before scaling')
sns.kdeplot(X_train['Age'], ax = ax1)

ax2.set_title('Age distribution after scaling')
sns.kdeplot(X_train_scaled['Age'], ax = ax2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12,5))

ax1.set_title('Estimated Salary distribution before scaling')
sns.kdeplot(X_train['EstimatedSalary'], ax = ax1)

ax2.set_title('Estimated salary distribution after scaling')
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax = ax2)
```

## Why Scaling is important??

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
lr_scaled.fit(X_train_scaled, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Actual accuracy score: ', accuracy_score(y_test, y_pred))
print('Scaled accuray: ', accuracy_score(y_test, y_pred_scaled))
```

### Desicion tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt_scaled = DecisionTreeClassifier()
```

```python
dt.fit(X_train, y_train)
dt_scaled.fit(X_train_scaled, y_train)
```

```python
y_pred = dt.predict(X_test)
y_pred_scaled = dt_scaled.predict(X_test_scaled)
```

```python
print('Actual : ', accuracy_score(y_test, y_pred))
print('Decision tree: ', accuracy_score(y_test, y_pred_scaled))
```

## Effect of Outlier

```python
new_data = pd.DataFrame({'Age': [5, 90, 95], 'EstimatedSalary': [1000, 250000, 350000], 'Purchased': [0, 1, 1]})
new_df = pd.concat([df1, new_data], ignore_index = True)
new_df = new_df.drop_duplicates(keep = 'first')
new_df
```

```python
plt.scatter(new_df['Age'], new_df['EstimatedSalary'])
#sns.scatterplot(x = df1['Age'], y= df1['EstimatedSalary'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(new_df.drop('Purchased', axis = 1),
                                                    new_df['Purchased'],
                                                    test_size = 0.3,
                                                    random_state = 0)

X_train.shape, X_test.shape
                                                    
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
plt.show()
```
