```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tips.csv')
df.sample(10)
```

### Information of data columns

```python
df.info()
```

### visualizing the categorical data

```python

sns.countplot(x = df['day'])
df['day'].value_counts()
```

```python
sns.countplot(x = df['time'])
df['time'].value_counts().index
```

```python
plt.pie(df['day'].value_counts(), labels = df['day'].value_counts().index, autopct = '%.2f%%')
plt.title('Pie chart')
plt.show()
```

###  For Numerical data

```python
plt.hist(x = df['tip'])
```

```python
sns.displot(df['total_bill'], kind = 'kde')
df['total_bill'].skew()
```

```python
df.columns
sns.boxplot(df['tip'], orient = 'h')
```

```python
plt.boxplot(df['tip'], vert = False)
```

```python
q1 , q3 = np.percentile(df['tip'], [25, 75])

iqr = q3 - q1
lower_boundary = q1 - 1.5 * iqr
upper_boundary = q3 + 1.5 * iqr
print('Lower boundary : ', lower_boundary)
print('Upper boundary : ', upper_boundary)

df1  = df[(df['tip']>=lower_boundary) & (df['tip'] <= upper_boundary)]
df1
```
